```python
from sklearn.preprocessing import LabelEncoder

# Assuming df is your DataFrame and le is an instance of LabelEncoder
cols = ['Education', 'EmploymentType', 'MaritalStatus', 'HasMortgage', 'HasDependents', 'LoanPurpose', 'HasCoSigner']
le = LabelEncoder()

for col in cols:
    df[col] = le.fit_transform(df[col])

df = df.drop('LoanID', axis=1)
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df['Default']
Y= df.drop('Default', axis=1)
```

```python
x_train, x_test, y_train, y_test= train_test_split(Y,X, test_size=0.2, random_state= 100)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from catboost import CatBoostClassifier
from sklearn.metrics import accuracy_score
```

```python
models=[
    LogisticRegression(max_iter=1000),
    RandomForestClassifier(),
    XGBClassifier(),
    KNeighborsClassifier(),
    GaussianNB(),
    CatBoostClassifier(verbose=False)
]
for model in models:
    model.fit(x_train, y_train)
    print(model.__class__.__name__)
    print ('train accuracy: ', accuracy_score(model.predict(x_train), y_train))
    print ('test accuracy: ', accuracy_score(model.predict(x_test), y_test))
```
