# combine the 2 datasets

```python
red['type']=1
white['type']=0
wines=pd.concat([white, red], ignore_index= True)
```

# Machin Learning

```python
from sklearn.model_selection import train_test_split
x= wines.drop('type', axis=1)
y=wines['type']
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report
```

```python
models = {
    "Logistic Regression": LogisticRegression(),
    "SVC": SVC(),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=100, random_state=100)
}

for model_name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc_score = accuracy_score(y_test, y_pred)
    print(f"{model_name}: Accuracy Score = {acc_score}")
    conf_matrix=confusion_matrix(y_pred,y_test)
    plt.figure(figsize=(5,5))
    sns.heatmap(conf_matrix, annot=True, cmap= 'coolwarm', xticklabels=['white','red'], yticklabels=['white','red'], fmt='d')
    plt.title('Confusion Matrix')
    plt.show()
    print('Classification report: ', classification_report(y_pred, y_test))
    
```

# Deep Learning

```python
from keras.models import Sequential
from keras.layers import Dense
model= Sequential()
model.add(Dense(12, activation='relu', input_shape=(12, )))
model.add(Dense(20, activation='relu'))
model.add(Dense(1, activation= 'sigmoid'))
model.output_shape
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics= ['accuracy'])
```

```python
model.fit(x_train,y_train, epochs=20, batch_size=8, verbose=1)
```

```python
# Evaluate the model on the test set
loss, accuracy = model.evaluate(x_test, y_test)

# Print out the accuracy
print("Accuracy on test set:", accuracy)
```

```python
import numpy as np

# Predict classes on the test set
y_pred = np.argmax(model.predict(x_test), axis=1)

# Compute confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(5,5))
sns.heatmap(conf_matrix, cmap='coolwarm', fmt='d', annot= True)
```
